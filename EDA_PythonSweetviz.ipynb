# 3. Side-by-Side Comparison of Train vs Test Set

```python
#separating X and y
X = penguins.drop('species', axis=1)
y = penguins['species']
```

## 3.1 Data Split using 80/20 Split Ratio

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
pd.set_option ('display.max_rows', 5)

X_train  #80% of dataset
```

```python
X_test  #20% of dataset
```

## 3.2 Comparison Report

```python
import sweetviz as sv
compare_report = sv.compare([X_train, 'Train'], [X_test, 'Test'])  #to compare 2 datasets (test and train in this case)
compare_report.show_html('Compare.html', open_browser = False)
```
