# Data Splitting in Python

## First import the data set we'll use.

```python
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("camnugent/california-housing-prices")
```

```python
ca_housing = pd.read_csv('/content/housing.csv')
```

```python
ca_housing
```

## Create X dataframe with predictor variables and y dataframe with outcome variable

### Choose predictor variables by index location and concatenate

```python
df1 = ca_housing.iloc[:,:-2]
df2 = ca_housing.iloc[:,-1:]

X = pd.concat ([df1,df2], axis=1)
X
```

### Choose outcome variable by column name

```python
y = ca_housing["median_house_value"]
y
```

## Use Scikit-learn to split the data

```python
from sklearn.model_selection import train_test_split
train_test_split (X, y, test_size = 0.2)  #the results will appear in the output
```

```python
#Create variables to capture the output
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.2)
```

### Add optional parameter to set a seed

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.2, random_state = 1337)
```
